```
def get_colour():

  current_list = [char for char in input()]
  colours = {"R", "G", "B"}
  
  while len(current_list) > 1:
    new_list = []
    for i in range(len(current_list)-1):
      if current_list[i] == current_list[i+1]:
        new_list.append(current_list[i])
      else:
        new_list.append(list(colours - {current_list[i], current_list[i+1]})[0])
    current_list = new_list

  return current_list[0]
```

```
def dots_boxes():

  def get_coords(number):
    return (number//6, number%6)

  def fill_boxes(boxes, grid, character):

    def is_box(grid, pos):
      return grid[pos[0]][pos[1]][0] and grid[pos[0]][pos[1]][1] and grid[pos[0]-1][pos[1]][0] and grid[pos[0]][pos[1]-1][1]

    for x in range(5):
      for y in range(5):
        if is_box(grid, (y, x)) and boxes[y][x] == "*":
          boxes[y][x] = character

  def place(grid, current_pos, jump):

    def get_direction(grid, pos):

      if grid[pos[0]][pos[1]][0] == 0 and pos[0] != 0:
        return 0
      if grid[pos[0]][pos[1]][1] == 0 and pos[1] != 5:
        return 1

      try:
        if grid[pos[0]-1][pos[1]][0] == 0:
          return 2
      except:
        pass

      try:
        if grid[pos[0]][pos[1]-1][1] == 0:
          return 3
      except:
        pass
      return 4

    first_pos = current_pos

    while get_direction(grid, get_coords(current_pos-1)) == 4:

      current_pos += jump
      current_pos %= 35
      if current_pos == first_pos:
        return 0

    coords = get_coords(current_pos-1)
    direction = get_direction(grid, coords)
    if direction == 0: grid[coords[0]][coords[1]][0] = 1
    elif direction == 1: grid[coords[0]][coords[1]][1] = 1
    elif direction == 2: grid[coords[0]-1][coords[1]][0] = 1
    else: grid[coords[0]][coords[1]-1][1] = 1

    return current_pos

  p_1, m_1, p_2, m_2, t = [int(i) for i in "4 10 14 23 47".split()]

  grid = [[[0, 0] for __ in range(6)] for __ in range(6)]  # [up, left]
  boxes = [["*" for __ in range(5)] for __ in range(5)]

  for i in range(t):
    p_1 = place(grid, p_1, m_1)
    fill_boxes(boxes, grid, "x")
    p_2 = place(grid, p_2, m_2)
    fill_boxes(boxes, grid, "o")
    
    if sum([k for j in (k for j in grid for k in j) for k in j]) == 60:
      break

    for row in grid:
      print(" ".join([("|" if cell[0] == 1 else ".")+("_" if cell[1] == 1 else " ") for cell in row]))

    for row in boxes:
      print(" ".join(row))
    print("\n")

  for row in boxes:
    print(" ".join(row))

  p_1_count = p_2_count = 0
  for row in boxes:
    for box in row:
      if box == "x":
        p_1_count += 1
      if box == "o":
        p_2_count += 1

  print("\n", p_1_count, p_2_count)
```
